```python
# Author: Aliya Rehman 
# MSc. Health Informatics
# City, University of London

# This script is being submitted as a part of "Knowledge Management in Healthcare" 2018. 
#Pre-processing Breast Cancer Data for cancer diagnosis using Pandas and SciPy

# import necessary libraries
import numpy as np
import pandas as pd
from scipy import stats

# read the original dataset into a PANDAS dataframe
df1 = pd.read_csv("BreastCancerData.csv")

# drop the last column as it contains all NAn values 
df1.drop(df1.columns[len(df1.columns)-1], axis=1, inplace=True)

# print data head for review
print(df1.head)
```

```python
# Drop all rows that have values of attributes farther than 3standard deviation away
df2 = df1[(np.abs(stats.zscore(df1.iloc[ :, 2:-1])) < 3).all(axis=1)]
df2.head
```

```python
# Outlier Removal in above step results as 495 examples
# Save the resulting dataset as a CSV file
df2.to_csv("BreastCancerData_NoOutliers.csv")
```
